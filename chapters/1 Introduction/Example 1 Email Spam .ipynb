```python
import pandas as pd
```

```python
df_spam = pd.read_csv("../../data/Spam.txt")
df_spam.head()
```

```python
print(df_spam.columns)
```

```python
words_feature = [word for word in df_spam.columns if 'word_freq' in word]
words = [word[10:] for word in words_feature]
# print(words_feature)
print(words)
```

```python
temp = df_spam[words_feature + ['spam']].groupby('spam').mean().sort_index(ascending=False)
temp.columns = words
temp.index = ['spam','email']
with pd.option_context("float_format",'{:,.2f}'.format):
    display(temp[(temp.iloc[0] - temp.iloc[1]).abs().sort_values(ascending=False).index[:12]])
```
